# India was the country with the highest death count

```python
cholera['proportion_of_deaths'] = (cholera.reported_deaths/cholera.reported_cases)
```

```python
cholera[(cholera['reported_deaths'] != 0.0)].dropna()
```

```python
cholera_50 = cholera[(cholera['Year'] >= 1950) & (cholera['Year'] <1960) & (cholera['reported_deaths'] != 0.0)].dropna()
```

```python
cholera_60 = cholera[(cholera['Year'] >= 1960) & (cholera['Year'] <1970) & (cholera['reported_deaths'] != 0.0)].dropna()
                                                   
```

```python
cholera_70 = cholera[(cholera['Year'] >=1970) & (cholera['Year'] < 1980) & (cholera['reported_deaths'] != 0.0)].dropna()
```

```python
cholera_80 = cholera[(cholera['Year'] >=1980) & (cholera['Year'] < 1990) & (cholera['reported_deaths'] != 0.0)].dropna()
```

```python
cholera_90 = cholera[(cholera['Year'] >=1990) & (cholera['Year'] < 2000) & (cholera['reported_deaths'] != 0.0)].dropna()
```

```python
cholera_2000= cholera[(cholera['Year'] >=2000) & (cholera['Year'] < 2015) & (cholera['reported_deaths'] != 0.0)].dropna()
```

```python
cholera_2000.head()
```

```python

ax = cholera_2000.plot(x = 'Year', y = 'proportion_of_deaths', label = 'India', c ='red')
ax.set_ylabel('proportion_of_death')
ax = cholera_50.plot(x = 'Year', y = 'proportion_of_deaths', label = 'Spain')
ax.set_ylabel('proportion_of_death')
```
